```python
import pandas as pd
```

```python
dados = pd.read_csv(r'manutencao_preditiva.csv')
dados.info()
```

```python
import matplotlib.pyplot as plt

x = dados['UDI']
y1 = dados['Temperatura Ar [K]']
y2 = dados['Temperatura Processo [K]']

plt.plot(x,y1, x, y2)
plt.xlabel("Quantidade")
plt.ylabel("Temperaturas")
plt.legend(["Temp. Ar", "Temp.Processo"])
plt.title("Curva de temperaturas x quantidade")
plt.show()
```

```python
#1. Mostre os valores mnimos, mximos e mdios da cada varivel numrica.
dados['UDI'].min()
dados['UDI'].mean()
dados['UDI'].max()
```

```python
#2. Mostre a quantidade de valores por tipo de mquina e por tipo de falha.

print(dados['Tipo'].value_counts())
print(dados['Tipo da Falha'].value_counts())


```

```python
#3. Mostre um grfico com as varivel UDI no eixo x e velocidade de rotao e torque o eixo y.
x = dados['UDI']
y1 = dados['Velocidade Rotacao [rpm]']
y2 = dados['Torque [Nm]']

plt.plot(x,y1,x,y2)
plt.xlabel("Quantidade")
plt.ylabel("Velocidade")
plt.legend(["Vel. Rotao", "Vel. Torque"])
plt.title("UDI x RPM/TORQUE")
plt.show()
```
